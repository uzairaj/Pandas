```python
import requests
import pandas as pd
download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
target_csv_path = "nba_data.csv"

response = requests.get(download_url)
response.raise_for_status()    # Check that the request was successful
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")


#When you execute the script, it will save the file nba_data.csv in your current working directory.
```

# Drop_Duplicates Function

drop_duplicates() removes duplicate rows from the dataset. This is a very important function when the dataset contains a lot of duplicate values.


Inplace=True implies that the changes will be made to the original dataset.



```python
print(len(nba_data))
nba_data.drop_duplicates(inplace=True)
print(len(nba_data))
```
